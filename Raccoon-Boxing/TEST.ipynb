```python
from imageai.Detection import ObjectDetection
from keras.preprocessing import image
from matplotlib import pyplot as plt
from boxing import draw_boxes
import os
```

```python
detector = ObjectDetection()
detector.setModelTypeAsYOLOv3()
detector.setModelPath(os.path.join(os.getcwd() , 'DATA/model/yolo.h5'))
detector.loadModel()
```

```python
img = image.load_img('DATA/test/person.jpg')
img = image.img_to_array(img)
```

```python
_, boxes = detector.detectObjectsFromImage(input_type='array', input_image=img, output_type='array',
                                                minimum_percentage_probability=50)
```

```python
bbs = []
for box in boxes:
    bb = dict()
    bb['xmin'] = int(box['box_points'][0])
    bb['ymin'] = int(box['box_points'][1])
    bb['xmax'] = int(box['box_points'][2])
    bb['ymax'] = int(box['box_points'][3])
    bb['class'] = box['name']
    bb['prob'] = box['percentage_probability']
    bbs.append(bb)
```

```python
def draw_boxes(img, boxes, show_labels=True):
    import cv2
    import numpy as np
    scale = 1e-3 * img.shape[0]
    line_t = int(img.shape[0]/800.)+1
    
    for box in boxes:
        #color = tuple(int(i) for i in np.random.choice(range(256), size=3))
        #text_color = tuple(int(255-i) for i in color)
        color = (0,255,0)
        text_color = (0,0,0)
        if show_labels:
            ann = box['class']+' '+str(box['prob'])[:5] if 'prob' in box else box['class']
            x_offset = ((len(ann)-1)*15.6+40)*scale+line_t*2
            y_offset = 40 * scale
            pts = np.array([[[box['xmin']-line_t+1, box['ymin']-y_offset],
                             [box['xmin']+x_offset, box['ymin']-y_offset],
                             [box['xmin']+x_offset, box['ymin']],
                             [box['xmin']-line_t+1, box['ymin']]]], dtype=np.int32)
            img = cv2.fillPoly(img, pts, color)
            cv2.putText(img,
                        ann,
                        (box['xmin'], box['ymin'] - 5),
                        16,
                        1e-3 * img.shape[0],
                        (0,0,0),
                        line_t,
                        cv2.LINE_AA)
        img = cv2.rectangle(img, (box['xmin'], box['ymin']), (box['xmax'], box['ymax']), color, line_t)
        #draw = ImageDraw.Draw(Image.fromarray(np.asarray(img, dtype="uint8")))
        #draw.text((box['xmin'], box['ymin'] - 6), ann, font=font)
    return img.astype('uint8')
```

```python
outImg = draw_boxes(img=img, boxes=bbs)
```

```python
plt.figure(figsize=(12,12))
plt.imshow(outImg)
plt.show()
```
